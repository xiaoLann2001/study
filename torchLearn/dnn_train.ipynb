```python
# 深度学习，最优化问题以损失函数最小化为目标
# 梯度下降法（最速下降法）：以当前点全局梯度反方向按指定步长搜索局部最优解，需要使用所有的样本（使样本损失函数最小的参数集合）
# 随机梯度下降算法：每次随机取一批样本(batch)计算梯度进行优化，一般为2的幂次方，缺点是更容易收敛到局部最优解，且难以跳出局部最优
```

```python
# 引入动量思想来优化模型，在每次更新参数时，在一定程度上会考虑之前参数更新的方向，同时利用当前batch计算得到的梯度，将两者结合得到最终需要
# 更新的大小和方向
```

```python
# 在训练开始时离最优值点较远，采用较大的学习率加快，几轮优化后需要使用较小的学习率减小振荡，精确搜索
# 对于训练的更新梯度大小和方向，以及自适应学习率有一系列的算法
```

```python
# 参数初始化torch.nn.init
import torch
# 1.针对某一层的权重进行初始化
conv1 = torch.nn.Conv2d(3, 16, 3)
torch.manual_seed(12)
# 使用标准正态分布初始化权重
torch.nn.init.normal_(conv1.weight, mean=0, std=1)
# 可视化
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
plt.hist(conv1.weight.data.numpy().reshape((-1, 1)), bins=30)
plt.show()
```

```python
# 使用指定值初始化偏置
torch.nn.init.constant_(conv1.bias, val=0.1)
```

```python
# 2.针对一个网络的权重初始化方法
class TestNet(nn.Module):
    def __init__(self):
        super(TestNet, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 3)
        self.hidden = nn.Sequential(
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 50),
            nn.ReLU(),
        )
        self.cla = nn.Linear(50, 10)
    def forward(self, x):
        x = self.conv1(x)
        x = x.view(x.shape[0], -1)
        x = self.hidden(x)
        output = self.cla(x)
        return output

testnet = TestNet()
print(testnet)
```

```python
# 获取所有参数
params = list(testnet.parameters())

# 绘制权重和偏置的分布情况
fig, axs = plt.subplots(len(params), 1, figsize=(3, 10))

for i, param in enumerate(params):
    # 将参数展平为一维数组
    param_data = param.data.cpu().numpy().flatten()
    
    # 绘制直方图
    axs[i].hist(param_data, bins=50)
    axs[i].set_title(f'Layer {i+1} Parameter Distribution')

plt.tight_layout()
plt.show()
```

```python
# 定义为网络中每个层进行权重初始化的函数
def init_weights(m):
    # 如果是卷积层
    if type(m) == nn.Conv2d:
        torch.nn.init.normal_(m.weight, mean=0, std=0.5)
    # 如果是全连接层
    if type(m) == nn.Linear:
        torch.nn.init.uniform_(m.weight, a=-0.1, b=0.1)
        m.bias.data.fill_(0.01)
```

```python
torch.manual_seed(13)
testnet.apply(init_weights)
```

```python
# pytorch中定义网络的方式
import torch
import torch.nn as nn
from torch.optim import SGD
import torch.utils.data as Data
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 读取数据
calif_X, calif_y = fetch_california_housing(return_X_y=True)
print("calif_X.shape: ", calif_X.shape)
plt.figure()
plt.hist(calif_y, bins=20)
plt.show()
```

```python
# 数据标准化处理
ss = StandardScaler(with_mean=True, with_std=True)
calif_Xs = ss.fit_transform(calif_X)
# 将数据预处理成可以使用PyTorch进行批量训练的格式
# 训练集x转化为张量
train_xt = torch.from_numpy(calif_Xs.astype(np.float32))
# 训练集y转化为张量
train_yt = torch.from_numpy(calif_y.astype(np.float32))
# 将训练集转化为张量后，使用TensorDataset将X和Y整理到一起
train_data = Data.TensorDataset(train_xt, train_yt)
# 定义一个数据加载器，将训练的数据集进行批量处理
train_loader = Data.DataLoader(
    dataset = train_data,   # 使用的数据集 
    batch_size = 256,       # 批处理样本大小
    shuffle = True,         # 每次迭代前打乱数据
    num_workers = 1,        # 使用两个进程
)
```

```python
# 使用nn.Sequential的形式
class MLPmodel2(nn.Module):
    def __init__(self):
        super(MLPmodel2, self).__init__()
        self.hidden = nn.Sequential(
            nn.Linear(8, 10),
            nn.ReLU(),
            nn.Linear(10, 10),
            nn.ReLU(),
        )
        self.regression = nn.Linear(10, 1)
    def forward(self, x):
        x = self.hidden(x)
        output = self.regression(x)
        return output
```

```python
mlp2 = MLPmodel2()
print(mlp2)
```

```python
# 对回归模型进行训练并输出损失函数的变化情况
optimizer = SGD(mlp2.parameters(), lr=0.001)
loss_func = nn.MSELoss()
train_loss_all = []

for epoch in range(30):
    for step, (c_x, c_y) in enumerate(train_loader):
        output = mlp2(c_x).flatten()
        train_loss = loss_func(output, c_y)
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()
        train_loss_all.append(train_loss.item())

plt.figure()
plt.plot(train_loss_all, "r-")
plt.title("Train loss per iteration")
plt.show()
```
